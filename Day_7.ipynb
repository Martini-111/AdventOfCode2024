```python
rawLines = [x.strip() for x in rawLines]

RHS_PAT = re.compile("(\s+\d+)")
LHS_PAT = re.compile("(\d+):")

RHSlist = list()
LHSlist = list()
for line in rawLines:
    RHSlist.append([int(num) for num in RHS_PAT.findall(line)])
    LHSlist.append(int(LHS_PAT.findall(line)[0]))

# for e1, e2 in zip(LHSlist, RHSlist):
#     # if(e1 == e2[-1]):
#     #     print( " MOTHERFUCKER ")
#     print(f"{e1}: {e2}")
totalCalibrationResult = 0

badCalibrationResult = 0
```

```python
class TreeNode:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None
    
    # def isLeaf(self):
    #     return (self.left==None and self.right==None)
    
    def isRightSkew(self):
        return (self.right!=None and self.left==None)
    
    def isLeftSkew(self):
        return (self.left!=None and self.right==None)
    
    def isStem(self):
        return (self.left!=None and self.right!=None)

def addNode(nodeVal, atNode, LHSval):
    global totalCalibrationResult
    global badCalibrationResult
    global calibrated
    global END_OF_TREE
    if atNode.isStem():
        # print("  /\\", end="\t")
        addNode(nodeVal, atNode.right, LHSval)
        addNode(nodeVal, atNode.left, LHSval)

    elif atNode.isRightSkew():
        print("  \\ RIGHTSKEW")
        addNode(nodeVal, atNode.right, LHSval)

    elif atNode.isLeftSkew():
        print("  / LEFTSKEW")
        addNode(nodeVal, atNode.left, LHSval)

    else:
        rightVal = nodeVal * atNode.data
        if rightVal < LHSval:
            atNode.right = TreeNode(rightVal)
        elif(rightVal > LHSval):
            print(f"---RIGHT CUT--- {rightVal} > {LHSval}")
        else:
            badCalibrationResult += LHSval
            print("\t\t\tIncoming Fireworks")

        leftVal = nodeVal + atNode.data
        if leftVal < LHSval:
            atNode.left = TreeNode(leftVal)
        elif(leftVal > LHSval):
            print(f"{leftVal} > {LHSval} ---LEFT CUT---")
        else:
            badCalibrationResult += LHSval
            print("\t\t\tIncoming Fireworks")
        
        if (leftVal == LHSval or rightVal == LHSval) and END_OF_TREE and not calibrated:
            totalCalibrationResult += LHSval
            print(f"\tTotal UPDATED {totalCalibrationResult-LHSval} to {totalCalibrationResult} by {LHSval}")
            calibrated = True
```

```python
calibSum = 0

class Node:
    def __init__(self, value, cause):
        self.value = value
        self.cause = cause # The value that led to creation
        self.left = None
        self.right = None
        # self.explored = False
    
```

```python
def include(node, number):
    global wanted
    global finalNum
    global calibSum

    if node.left != None and node.right != None:
        include(node.left, number)
        include(node.right, number)

    elif node.left == None and node.right != None:
        include(node.right, number)

    elif node.left != None and node.right == None:
        include(node.left, number)

    else:
        leftnum = node.value + number
        rightnum = node.value * number
        node.left = Node(leftnum, number)
        node.right = Node(rightnum, number)

        if finalNum==True and (leftnum==wanted or rightnum==wanted):
            finalNum = False
            calibSum += wanted
```
