```python
import re
from bisect import bisect_left, bisect_right
```

```python
fobj  = open("Day_5_input.txt", "r")
inpStr = fobj.read()
# inpStr
```

```python
pattern = re.compile("(\d{1,2})\|(\d{1,2})")
orderRulesStr = pattern.findall(inpStr)
# print(orderRulesStr, "\n")

pos = inpStr.find(orderRulesStr[-1][0]+"|"+orderRulesStr[-1][1]) # This could have been fucked by a rule repetition of the final GUY
pos += len(orderRulesStr[-1][0]+"|"+orderRulesStr[-1][1]+"\n") + 1 # Another newline character
orderRules = [(int(x), int(y)) for (x,y) in orderRulesStr]

# print(inpStr[pos+7:pos+20])
print(orderRules[-1])


inpList = [x.split(",") for x in inpStr[pos:].split("\n")]
inpList = [[int(x) for x in y] for y in inpList]

print(inpList[0])
```

```python
orderRules = sorted(orderRules, key=lambda tup: tup[0]+tup[1])
orderRules = sorted(orderRules, key=lambda tup: tup[1])
gkeys = [tup[1] for tup in orderRules]
lkeys = [tup[0] for tup in orderRules]
# orderRules
```

```python
def middle_man(intlist):
    return intlist[int(len(intlist)/2)]

def swap(inpList, ind1, ind2, coolbool=False):
    temp = inpList[ind2]
    inpList[ind2] = inpList[ind1]
    inpList[ind1] = temp
    if coolbool:
        print(f"swapped {temp} and {inpList[ind2]}")
    
```

```python
badrowcount = 0
badrowList = []
for inpRow in inpList[:]:
    for i in range(len(inpRow)):
        valid = True

        # Searching for rules that places inpRow[i] as GREATER
        startInd = bisect_left(gkeys, inpRow[i])
        stopInd = startInd + bisect_right(gkeys[startInd:], inpRow[i])

        if(startInd<0 or stopInd>len(orderRules)):
            raise KeyError("arrayOutOfBoundsException")
        # print(orderRules[startInd])
        # print(orderRules[stopInd])
        for j in range(i+1, len(inpRow)):
            if inpRow[j] in lkeys[startInd:stopInd]:
                print(f"This row is invalid: {inpRow[:j].__str__()} !!{inpRow[j]} </{inpRow[i]}>!! {inpRow[j+1:].__str__()}")
                badrowcount += 1
                valid=False
                break
        if valid == False:
            badrowList.append(inpRow)
            inpList.remove(inpRow)
            break
```
